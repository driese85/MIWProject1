```python
import requests
import pandas as pd
import json
import time
import csv
```

```python
# First url to make call to
url = 'http://localhost:8080/medish_centrum_randstad/api/netlify?page=1'
# Initialize empty list to store data from API calls
data = []

# create infinite loop to make requests to API
while True:
    # log which call is being made
    print(f'Making call to: {url}')
    # Make call
    r = requests.get(url)
    # if call is unsuccessfull (status code not equal to 200), then try again
    if r.status_code != 200:
        continue
    # check if content we expect to receive JSON, is actually in JSON format
    if r.headers['Content-Type'] == 'application/json':
        # using library "json" to transform text string to dictionary in python
        data_json = json.loads(r.text)
        # extend resulting list of dictionaries to data list created at the top
        print(len(data_json))
        data.extend(data_json['data'])
    # implement pagination, check if the call to the API returns a "next" attribute
    if 'next' in data_json:
        # get the url from the next attribute and save it to the existing url variable for use in the next loop operation
        url = data_json['next']
        # sleep for 300 ms to prevent from being blacklisted by the API server
        time.sleep(0.3)
    # if the API call does not return a "next" attribute, we have succesfully retrieved all data from the server and we can exit the while loop
    else:
        break

print(len(data))
with open('../data/data.csv', 'w', newline='') as csvfile:
    fieldnames = ['genetic', 'length', 'mass', 'exercise', 'smoking', 'alcohol', 'sugar', 'lifespan']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for row in data:
        writer.writerow(row)
            


```
